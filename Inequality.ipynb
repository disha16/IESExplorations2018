```python
import pandas as pd
import statsmodels.api as sm
import scipy
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.pylab as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_pacf
df2 = pd.read_csv("C:\Users\Rawal Rawal\Downloads\Tab98.csv",thousands=',')
df2.head()
```

```python
df3 = df2[df2['Indicators'].isin(["Urban Average MPCE 2009-10 (in `)","Rural Average MPCE 2009-10 (in `)","Education Index 2007-08"])]
df3.index = df3['Indicators']
del df3['Indicators']
df31 = df3.T
df31['Urban Average MPCE 2009-10 (in `)']= df31['Urban Average MPCE 2009-10 (in `)'].apply(float)
df31['Rural Average MPCE 2009-10 (in `)']= df31['Rural Average MPCE 2009-10 (in `)'].apply(float)
df31['Diff'] = ((df31['Urban Average MPCE 2009-10 (in `)']- df31['Rural Average MPCE 2009-10 (in `)'])/df31['Rural Average MPCE 2009-10 (in `)'])
df31['Diff']= df31['Diff']*100

del df31["Rural Average MPCE 2009-10 (in `)"]
del df31["Urban Average MPCE 2009-10 (in `)"]
p1 = sns.regplot(data=df31, x="Education Index 2007-08", y="Diff", fit_reg=False, marker="+", color="skyblue")


```
